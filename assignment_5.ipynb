Import the library pandas and load the airports.csv data to the program.

```python
import pandas as pd
```

```python
ap = pd.read_csv("https://raw.githubusercontent.com/tidyverse/nycflights13/master/data-raw/airports.csv")
```

#1. What is the northernmost airport in the United States? 

The northern most airport in United States has the highest latitude.

```python
ap.head()
ap.sort_values("lat", ascending=False).head(5)
```

The northern most airport in the United States is Dillant Hopkins Airport with a latitude of 72.270833. 

#2. What is the easternmost airport in the United States?

The airports on the eastern most part of United States is located between the latitude of 44 to 50 and longitude of -65 and -75.

```python
n = ap[(ap['lat'] > 44) & (ap['lat'] <= 50) & (ap['lon'] >= -75) & (ap['lon'] <= -65)]
n.sort_values('lat', ascending=False).head(5)
```

The eastern most airport in the United States is Northern Aroostook Regional Airport. It has a latitude of of 47.285556 and a longitude of -68.312778

#3. On February 12th, 2013, which New York area airport had the windiest weather?

Load the weather.csv data into the program and display the head of the data.

```python
win = pd.read_csv("https://raw.githubusercontent.com/tidyverse/nycflights13/master/data-raw/weather.csv")
win.head(5)
```

Select the month to February represented by 2, day to be 12th and year to be 2013. The values for sorting is wind speed sorted in descending order.

```python
win[(win["month"]==2) & (win["day"]==12) & (win["year"]==2013)].sort_values("wind_speed", ascending=False).head(5)
```

The EWR is an outlier due to the abnormal wind speed of 1048.36 mph. As a result the windiest airport in New York area in February, 13 2013 is LGA with a 23.02 mph.
